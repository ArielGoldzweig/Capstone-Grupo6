```python
def calcular_correcciones_tramos(tramos, tramos_os):
  correcciones = list()

  for i in range(len(tramos) - 1):
    x = np.array(tramos[i])
    y = np.array(tramos_os[i])
    model = sm.RLM(y, x, M=sm.robust.norms.HuberT())
    results = model.fit()
    slope = results.params[0]
    correcciones.append(slope)

  return correcciones


def calcular_correccion(distancias_simulacion, distancias_os):

  x = np.array(distancias_simulacion)
  y = np.array(distancias_os)
  model = sm.RLM(y, x, M=sm.robust.norms.HuberT())
  results = model.fit()
  slope = results.params[0]

  return slope


```
